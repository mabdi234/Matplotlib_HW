```python
%matplotlib notebook
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
cfile = "city_data.csv"
rfile = "ride_data.csv"

# Read the City and Ride Data

city=pd.read_csv(cfile)
rides=pd.read_csv(rfile)



# Combine the data into a single dataset


combined=pd.merge(city,rides, on = ['city'])

combined

# Display the data table for preview
```

```python
#Getting data for plot


urban_data=combined.loc[combined['type'] == 'Urban'].drop(['type','ride_id','date'], axis=1)
suburban_data=combined.loc[combined['type'] == 'Suburban'].drop(['type','ride_id','date'], axis=1)
rural_data=combined.loc[combined['type'] == 'Rural'].drop(['type','ride_id','date'], axis=1)

urban_ride_count= pd.Series(urban_data['city']).value_counts()
suburban_ride_count=pd.Series(suburban_data['city']).value_counts()
rural_ride_count=pd.Series(rural_data['city']).value_counts()

urban_ride_df=pd.DataFrame(urban_ride_count)

urban_cities={"cities":[urban_ride_df.index]}
urban_df= pd.DataFrame(urban_cities)


#print(urban_df)

#merged_urban=pd.merge()

#print(urban_ride_df)

#Here is where is I do my averges

avg_urban=urban_data.groupby('city').mean().reset_index()
avg_suburban=suburban_data.groupby('city').mean().reset_index()
avg_rural=rural_data.groupby('city').mean().reset_index()

#print(suburban_ride_count)

print(avg_rural)

#urban_d=drivers.loc['Urban'].sum()['driver_count'].item()

#print(avg_urban)

urban_driver_count=urban_data.groupby(["city"]).mean()["driver_count"]
#print(urban_driver_count)
```
